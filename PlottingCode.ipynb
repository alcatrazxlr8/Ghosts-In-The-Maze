```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('a1.txt', sep=" ", header=None)
df.columns = ["x", "y"] #  , "y1", "y2", "y3","y4"
print(df)
```

```python
# AGENT-1
df1 = pd.read_csv('a1.txt', sep=" ", header=None)
df1.columns = ["x", "y"] #  , "y1", "y2", "y3","y4"
# print(df)
fig, ax = plt.subplots()
ax.plot(df1['y'],df1['x'], label='Agent-1')
# tick_spacing = 1
# ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.xlabel('Ghosts')
plt.ylabel('Survivability')
plt.title('Agent-1')
plt.legend('1')
plt.show()
```
